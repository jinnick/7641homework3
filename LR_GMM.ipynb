```python
import import_ipynb
import letter_recognition as data
import numpy as np
from sklearn import metrics
```

```python
from sklearn.mixture import GaussianMixture as GMM
ll1=[]
bic1=[]
clusters=range(1,31)
model = GMM(covariance_type = 'diag')
for k in clusters:
            model.set_params(n_components=k)
            model.fit(data.X_train)
            labels = model.predict(data.X_train)
            ll1.append(model.score(data.X_train))
            bic1.append(model.bic(data.X_train))
            print(model.score(data.X_train))
print(bic1)
print(len(ll1))
```

```python
from sklearn.mixture import GaussianMixture as GMM
ll2=[]
bic2=[]
clusters=range(1,31)
model = GMM(covariance_type = 'spherical')
for k in clusters:
            model.set_params(n_components=k)
            model.fit(data.X_train)
            labels = model.predict(data.X_train)
            ll2.append(model.score(data.X_train))
            bic2.append(model.bic(data.X_train))
            print(model.score(data.X_train))
print(bic2)
```

```python
from sklearn.mixture import GaussianMixture as GMM
ll3=[]
bic3=[]
clusters=range(1,31)
model = GMM(covariance_type = 'full')
for k in clusters:
            model.set_params(n_components=k)
            model.fit(data.X_train)
            labels = model.predict(data.X_train)
            ll3.append(model.score(data.X_train))
            bic3.append(model.bic(data.X_train))
            print(model.score(data.X_train))
print(bic3)
```

```python
from sklearn.mixture import GaussianMixture as GMM
ll4=[]
bic4=[]
clusters=range(1,31)
model = GMM(covariance_type = 'tied')
for k in clusters:
            model.set_params(n_components=k)
            model.fit(data.X_train)
            labels = model.predict(data.X_train)
            ll4.append(model.score(data.X_train))
            bic4.append(model.bic(data.X_train))
            print(model.score(data.X_train))
print(bic4)
```

```python
from sklearn import metrics
homogeneity_scores=[]
clusters=range(1,31)
for k in clusters:
    gmm = GMM(covariance_type = 'full')
    gmm.set_params(n_components=k)
    label = gmm.fit_predict(data.X_train)
    homogeneity_scores.append(metrics.homogeneity_score(data.y_train,label ))
plt.plot(clusters, homogeneity_scores,'bo-')
plt.xlabel('k(Number of clusters)',fontsize=18)
plt.ylabel('Homogeneity Score',fontsize=18)
plt.title('Letter_Recognition(GMM_Full_Covariance_Homogeneity Score)',fontsize=14)
plt.show()
```

```python
from sklearn import metrics
from sklearn.mixture import GaussianMixture as GMM
import matplotlib.pyplot as plt
completeness_scores=[]
clusters=range(1,31)
for k in clusters:
    gmm = GMM(covariance_type = 'full')
    gmm.set_params(n_components=k)
    label = gmm.fit_predict(data.X_train)
    completeness_scores.append(metrics.completeness_score(data.y_train,label ))
plt.plot(clusters, completeness_scores,'g^-')
plt.xlabel('k(Number of clusters)',fontsize=18)
plt.ylabel('Completeness Score',fontsize=18)
plt.title('Letter_Recognition(GMM_Full_Covariance_Completeness Score)',fontsize=14)
plt.show()
```
